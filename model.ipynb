```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib
matplotlib.rcParams.update({'font.size':18})
matplotlib.rcParams.update({'font.family':'serif'})
```

```python
def PEC(P, t=1):
    '''
    Estimate of the uncertainty in a period
    based on Kepler long-cadence data for eclipsing binary stars
    
    Matching scaling in Figure 5 from Mighell & Plavchan (2013),
    and using Eqn 27 for the error model.
    https://doi.org/10.1088/0004-6256/145/6/148
    
    P = period in days
    t = observing baseline in years
    
    '''
    log_sig = -5.8908 + 1.4425 * (1. + np.log10(P))
    
    # rescale to match Figure 5
    scale = (125 / 365 / t)**(3/2)
    
    return 10**log_sig * scale
```

```python
plt.figure(figsize=(7,6))
ttmp = 10**np.arange(-1, 3, 0.01)

plt.plot(ttmp, PEC(ttmp,t=1), label='1 yr of data', c='C0')
plt.plot(ttmp, PEC(ttmp,t=2), label='2 yr', c='C4')
plt.plot(ttmp, PEC(ttmp,t=4), label='4 yr', c='C3')
plt.plot(ttmp, PEC(ttmp,t=8), label='8 yr', c='C1')
plt.legend(fontsize=14)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Period (days)')
plt.ylabel('PEC Error (days)')
plt.title('Replicating Mighell & Plavchan (2013) FIG 5', fontsize=14)
```
